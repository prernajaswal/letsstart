```python
df["revenue"] = df["price"] * df["quantity_sold"]
```

```python
df
```

```python
df['final_revenue'] = df['revenue'] - (df['revenue'] * df['discount_percent'] / 100)
```

```python
df
```

```python
result = (
    df
    .groupby("category", as_index=False)
    .agg(total_final_revenue=("final_revenue", "sum"))
)

result
```
